# book_title, publisher, book_author 통일하기

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from dython.nominal import associations
```

```python
users = pd.read_csv('users-mod.csv')
ratings = pd.read_csv('ratings-mod.csv')
books = pd.read_csv('books-mod.csv')


JH_dir = '../junhyuk/'
m_books = pd.read_csv(JH_dir+'basic_preprocess_books.csv', index_col = 'Unnamed: 0')
```

```python
m_books
```

```python
m_books.groupby(['book_title','publisher'])['book_author']
```

```python
gr_ath = m_books.groupby(['book_title','publisher'])['book_author'].apply(','.join)
gr_ath = pd.DataFrame(gr_ath)
```

```python
gr_ath
```

```python
# gr_ath[gr_ath['book_author'].str.contains(',')]
gr_ath['book_author'] = gr_ath['book_author'].str.split(',')
gr_ath['book_author'] = gr_ath['book_author'].apply(lambda x: x[0])

gr_ath.head()
```

```python
m_books
```

```python
books_oneath = m_books.merge(gr_ath, on=['book_title', 'publisher'])
del books_oneath['book_author_x']

books_oneath = books_oneath.rename(columns={'book_author_y':'book_author'})
books_oneath.head()
```

```python
books_oneath
```

```python
books_oneath['category_high'] = books_oneath['category_high'].fillna('')
books_oneath
```

같은 book_title, publisher의 책들의 author 통일 완료.

## best-seller author 생성

```python
import pandas as pd
```

```python
books_oneath = pd.read_csv('books_united_authors.csv')
books_oneath
```

```python
author_count = books_oneath.groupby('book_author').count().sort_values('isbn', ascending=False)
author_count.describe()
```

100권 이상을 best_seller author로 지정. 나머지 author은 others로 바꿈

```python
# count 100 이상인 authors를 저장 -> 저장한 author_list에 없으면 모두 other로 바꿈
author_count[author_count['isbn'] >= 100].index.__len__()
```

```python
author_count[author_count['isbn'] >= 50].index.__len__()
```

```python
author_count[author_count['isbn'] >= 10].index.__len__()
```

```python
# 10권 이상으로 진행
authors_over10 = author_count[author_count['isbn'] >= 10].index
authors_over10 = authors_over10.drop('Not Applicable (Na )')  # na 명명된 거 뻄
authors_over10
```

```python
books_oneath['book_author_over10'] = books_oneath['book_author'].apply(lambda x: x if x in authors_over10 else 'others')
books_oneath
```

```python
books_oneath.groupby('book_author_over10').count()#.plot(y='isbn')
# over3 5 10 50 100 
```

```python
books_oneath.to_csv('books_oneauthor_over10.csv', index=None)
```
