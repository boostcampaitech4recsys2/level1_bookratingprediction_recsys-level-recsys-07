```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_csv("data/train_ratings.csv")
books = pd.read_csv("data/books_final.csv")
users = pd.read_csv("data/users_preprocessed.csv")
```

```python
data = ratings.merge(users, "left", 'user_id').merge(books, "left", 'isbn')
```

```python
data.head()
```

```python
# data['rating>5'] = 
def tmp(x):
    if x <= 5:
        return 0
    elif x <= 8:
        return 1
    else:
        return 2
data['rating_group'] = data['rating'].map(tmp)
```

```python
data = data.merge(data.groupby('user_id')['rating'].mean().rename('user_rating_mean').reset_index(), 'left', 'user_id')
data = data.merge(data.groupby('isbn')['rating'].mean().rename('book_rating_mean').reset_index(), 'left', 'isbn')
```

```python
data = data.merge(data.groupby('user_id')['isbn'].count().rename('user_book_count').reset_index(), 'left', 'user_id')
data = data.merge(data.groupby('isbn')['user_id'].count().rename('book_user_count').reset_index(), 'left', 'isbn')
```

```python
# new_language
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
tmp = data.groupby(['rating', 'new_language'])['isbn'].agg(['count']).sort_index().unstack()
tmp.apply(lambda x: x/tmp.apply(lambda x: x.sum(skipna=True), axis=1), axis=0).mul(100)
```

```python
# category_high
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
tmp = data.groupby(['rating', 'category_high'])['isbn'].agg(['count']).sort_index()
# tmp2 = tmp.apply(lambda x: x/tmp.apply(lambda x: x.sum(skipna=True), axis=1), axis=0).mul(100)
tmp
```
