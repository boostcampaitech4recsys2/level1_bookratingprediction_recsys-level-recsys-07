```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_csv("../data/train_ratings.csv")
books = pd.read_csv("../data/books_final.csv")
users = pd.read_csv("../data/users_preprocessed.csv")
```

```python
data = ratings.merge(users, "left", 'user_id').merge(books, "left", 'isbn')
```

```python
data.head()
```

```python
# 유저별 평균 평점 or 중앙값
user_rating_stat = ratings.groupby(['user_id'])['rating'].agg(['count','mean','median','std']).reset_index()
book_rating_stat = ratings.groupby(['isbn'])['rating'].agg(['count','mean','median','std']).reset_index()
```

```python
new_users = users.merge(user_rating_stat.fillna(0), "left", "user_id")
new_books = books.merge(book_rating_stat.fillna(0), "left", "isbn")
```

```python
new_users.to_csv("users_stat.csv", index=False)
new_books.to_csv("books_stat.csv", index=False)
```
